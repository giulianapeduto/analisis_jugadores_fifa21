```python
import pandas as pd

df = pd.read_csv("players_21.csv")

df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
top10 = df.sort_values(by="value_eur", ascending=False).head(10)

top10[["short_name","club_name","value_eur"]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.barplot(
    x="value_eur",
    y="short_name",
    data=top10
)

plt.title("Top 10 jugadores más valiosos FIFA 21")
plt.xlabel("Valor en euros")
plt.ylabel("Jugador")

plt.show()
```

```python
top_paises = df["nationality"].value_counts().head(10)

top_paises
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=top_paises.values,
    y=top_paises.index
)

plt.title("Top 10 países con más jugadores FIFA 21")
plt.xlabel("Cantidad de jugadores")
plt.ylabel("País")

plt.show()
```

```python
clubes_valor = df.groupby("club_name")["value_eur"].sum().sort_values(ascending=False).head(10)

clubes_valor
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=clubes_valor.values,
    y=clubes_valor.index
)

plt.title("Clubes con planteles más valiosos")
plt.xlabel("Valor total del plantel")
plt.ylabel("Club")

plt.show()
```

```python
df["main_position"] = df["player_positions"].str.split(",").str[0]

df[["short_name","player_positions","main_position"]].head()
```

```python
valor_posicion = df.groupby("main_position")["value_eur"].mean().sort_values(ascending=False)

valor_posicion.head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=valor_posicion.head(10).values,
    y=valor_posicion.head(10).index
)

plt.title("Valor promedio por posición")
plt.xlabel("Valor promedio en euros")
plt.ylabel("Posición")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x="age",
    y="value_eur",
    data=df
)

plt.title("Relación entre edad y valor de mercado de jugadores")
plt.xlabel("Edad")
plt.ylabel("Valor en euros")

plt.show()
```

```python
valor_por_edad = df.groupby("age")["value_eur"].mean()

valor_por_edad.head()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x=valor_por_edad.index,
    y=valor_por_edad.values
)

plt.title("Valor promedio de jugadores según la edad")
plt.xlabel("Edad")
plt.ylabel("Valor promedio en euros")

plt.show()
```
